```python
#Create  a Dog class with the following attributes: name (string), age (integer), breed (string)

#Dog Class
class Dog:
   def __init__(self, name, age, breed):
       self.name = name
       self.age = age
       self.breed = breed
       #Method __str__ to returna  string representation of the dog
   def __str__(self):
       return f"Name: {self.name}, Age: {self.age}, Breed: {self.breed}"


#Shelter Class
class Shelter:
   def __init__(self):
       """
       Initialize the Shelter with an empty list of dogs.
       """
       self.dogs = []


   def add_dog(self, dog):
       """
       Add a Dog object to the shelter.
       """


       self.dogs.append(dog)
       print(f"{dog.name} has been added to the shelter.")


   def remove_dog(self, name):
       """
       Remove a Dog object from the shelter by name.
       If the dog is not found, display a message.
       """


       for dog in self.dogs:
           if dog.name == name:
               self.dogs.remove(dog)
               print(f"{name} has been removed from the shelter.")
               return
       print(f"Dog named {name} was not found in the shelter.")


   def display_dogs(self):
       """
       Display all dogs in the shelter.
       """


       if not self.dogs:
           print("No dogs are currently in the shelter.")
       else:
           for dog in self.dogs:
               print(dog)


# Driver Code
if __name__ == "__main__":
   # Creating a few Dog objects
   dog1 = Dog("Buddy", 3, "Golden Retriever")
   dog2 = Dog("Max", 2, "German Shepherd")
   dog3 = Dog("Bella", 1, "Bulldog")


   # Create Shelter instance
   shelter = Shelter()


   # Add dogs to the shelter
   shelter.add_dog(dog1)
   shelter.add_dog(dog2)
   shelter.add_dog(dog3)


   # Display all dogs in the shelter
   print("All dogs in the shelter:")
   shelter.display_dogs()


   # Remove a dog from the shelter
   shelter.remove_dog("Max")


   # Display all dogs in the shelter after removal
   print("\nAfter removing Max:")
   shelter.display_dogs()


   # Attempt to remove a dog that doesn't exist
   shelter.remove_dog("Rocky")

```
